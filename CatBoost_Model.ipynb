```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier, Pool, metrics, cv
from sklearn.metrics import accuracy_score
```

```python
#importer les données de training et test
trainData = pd.read_csv('Data/train.csv')
trainLabels = pd.read_csv('Data/train_result.csv')
testData = pd.read_csv('Data/test.csv')
```

```python
#Mettre les points en bas de 0 à 0 et enlver les NaN
trainData[trainData < 0] = 0
trainData = trainData.dropna(axis=1)

testData[testData < 0] = 0
testData = testData[trainData.columns]
```

```python
#Créer les données de training et validation
trainX,validX,trainY,validY = train_test_split(trainData,trainLabels,test_size = 0.2, random_state = 1)
```

```python
#Entrainer le modèle CatBoost
model = CatBoostClassifier(custom_loss=[metrics.Accuracy()],random_seed=42)
model.fit(trainX, trainY['Class'],eval_set=(validX, validY['Class']),
        logging_level='Verbose',  # you can uncomment this for text output
        plot=True)
```

```python
#Faire les prédictions
predictions = model.predict(testData)
```

```python
#Transformer et exporter les prédictions
predict = pd.DataFrame([range(10000),predictions], index = ['Index','Class']).transpose()
predict['Class'] = predict['Class'].str.strip('[]').astype(int)

predict.to_csv('predictions_CatBoost.csv',index=False)
```
